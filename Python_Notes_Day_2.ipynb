## 1. For Loop

### Theory

The `for` loop in Python is used for iterating over a sequence (such as a list, tuple, dictionary, set, or string). This is different from traditional index-based `for` loops in C/Java. Python's `for` loop uses **iterators**, making it concise and readable.


### Syntax:

```python
for item in iterable:
    # block of code



````python
# Example 1: Looping through a list

names = ["Putin", "Xi Jinping", "Kim"]
for name in names:
    print(f"Hello, {name}!")
````

````python
# Looping through a string

for char in "Legend Shekokar (sir)":
    print(char.lower())
````

# 3. Loop Control Statements
### Python provides a few tools to control the flow inside loops.



### break: Stop the loop entirely

````python
for i in range(5):
    if i == 3:
        break
    print(i)
````

###continue: Skip to the next iteration of the loop




````python
txt = "rah-ul-ga-n-d-hi"

for i in txt:
    if i == "-":
        continue
    print(i, end="")
````

###pass: Do nothing (a placeholder)


````python
for i in range(3):
    if i == 1:
        pass
    print("Running", i)
````

##4. range() and enumerate()
###Theory
The range() function generates a sequence of numbers, often used to run a loop a specific number of times. enumerate() allows you to loop with both index and valueâ€”helpful for debugging and traceability.



### range() syntax


    range(start, stop, step):

    start: optional (default=0). inclusive.
    stop: required. exclusive.
    step: optional (default=1)

````python
list(range(0, 5, 2))
# [0, 1, 2, 3, 4]
# [-5,-4,-3,-2,-1]
````

````python
#1: Simple count

for i in range(5):
  print(i)
````

##5. try / except / else / finally (Error Handling)
###Theory
Pythonâ€™s try/except blocks help catch and handle errors without crashing the program.

```python
try: ---(required)
    # risky code
except SomeError(optional): ---(required)
    # recovery code
else: ---(optional)
    # runs only if no exception occurs
finally: ---(optional)
    # always runs


Example 1: ValueError

````python
try:
    gpa = int(input("Enter GPA: "))
    print("Your GPA:", gpa)
except:
    print("That wasn't a number!")
````

Example 2: Multiple except blocks


````python
try:
    num = int(input("Enter a number: "))
    print(10 / num)
except ValueError:
    print("Invalid input.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
````

Example 3: With else

````python
try:
    value = int(input("Enter a number: "))
except ValueError:
    print("Invalid input!")
else:
    print("Square is:", value ** 2)
````

Example 4: With finally

````python
try:
    print("Mathing...\n")
    result = 10 / 1
except ZeroDivisionError:
    print("Math error")
finally:
    print("Delulu is the solulu")
````
