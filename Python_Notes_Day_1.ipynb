# ***Indentation: The Most Important Rule!***

In Python, indentation is not just for making code look pretty—it's a rule that defines code structure. While other languages use curly braces {} to group code, Python uses whitespace.

**Real-life Analogy**: Think of an outline for an essay. The main points are on the left, and the sub-points are indented underneath them. This structure shows which ideas belong to which topic. In Python, indentation shows which lines of code belong to a specific block, like an if statement or a function.

**The Rule**: You must use consistent indentation, typically four spaces, for code blocks.

```python
# This is a comment. The Python interpreter ignores it.

# Let's define a variable (we'll cover this soon!)
is_raining = True

# The 'if' statement starts a new block of code.
# Notice the colon (:) at the end of the line.
if is_raining:
    # This line is INDENTED, so it only runs if the 'if' condition is true.
    print("Don't forget your umbrella!")
    print("And wear your boots!")

# This line is NOT indented, so it is not part of the 'if' block.
# It will run regardless of whether it's raining or not.
print("Have a great day!")
```

# ***🧠 Control Flow: Making Decisions***

So far, our code has run from top to bottom, one line at a time. But that's not how real life works. We make decisions constantly. Control flow statements allow our programs to do the same.

**Real-life Analogy:**
You're at a crossroads.

IF it's raining, you'll go to the mall.

ELSE IF (ELIF) it's sunny and warm, you'll go to the beach.

ELSE (if neither of the above is true), you'll just stay home. This is exactly how if, elif, and else work in Python. They control the "flow" of your program's execution.

**if, elif, else**

These statements let you run certain blocks of code only when a specific condition is True.

```python
a = 10
b = 20
if a>b :
  print(f"{a} is greater than {b}")
else :
  print(f"{b} is greater than {a}")
```

```python
'''
if some_condition:
    # Code to run if some_condition is True
elif another_condition:
    # Code to run if some_condition was False,
    # but another_condition is True
else:
    # Code to run if all previous conditions were False

'''
```

**Key Points:**

1. You start with if.

2. You can have zero or more elif statements.

3. You can have at most one else statement at the end.

4. The elif and else parts are optional.

5. Don't forget the colon (:) at the end of each condition line and the indentation for the code blocks!

```python
# Example 1: A simple if-else statement
age = int(input("Enter your age: "))

if age >= 18:
    print("You are eligible to vote.")
else:
    print("Sorry, you are not yet eligible to vote.")


```

```python
# Example 2: An if-elif-else ladder for grading
score = int(input("Enter your exam score (0-100): "))

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Your grade is: {grade}")

```

```python
# Example 3: A simple 'if' without an else
temperature = 25
if temperature > 30:
    print("It's a hot day! Stay hydrated.")

print("Weather check complete.")
```

# ***Nested Conditions***

You can put if statements inside other if statements. This is called **nesting**.

**Real-life Analogy**: You want to apply for a premium credit card. First, they check IF your credit score is high enough. IF it is, they then check IF your annual income is above a certain threshold. The second check is nested inside the first.

```python
# A theme park ride checker
age = int(input("Enter your age: "))
height_cm = int(input("Enter your height in cm: "))

if height_cm >= 140:
    print("You are tall enough to ride.")
    # This is a nested condition. It only gets checked if the first 'if' is True.
    if age >= 16:
        print("You can ride the 'Python's Plunge' roller coaster!")
    else:
        print("You can ride the 'Tamer Twister'.")
else:
    print("Sorry, you are not tall enough to ride this roller coaster.")
```

# ***🛠️ Functions: Reusable Code Blocks***

Imagine you have a set of instructions you need to perform over and over again. Instead of rewriting them every time, you can bundle them into a function.

**Real-life Analogy:** A recipe for making tea. The recipe is the function definition. It has a name ("make_tea") and lists the steps. Every time you want tea, you call the function ("I'm going to make tea now"). The ingredients you use (like sugar or milk) are the parameters.

**Why use functions?**

**1. DRY Principle**: Don't Repeat Yourself. It makes your code shorter and cleaner.

**2. Organization**: Breaks a big problem down into smaller, manageable pieces.

**3. Reusability**: Write once, use everywhere.

**Defining and Calling a Function**

```python
def func_name(a,b):
    return

func_name(10,20)
```

```python
'''

# DEFINING the function
def function_name(parameter1, parameter2):
    # Indented code block: the body of the function
    # Do stuff with parameter1 and parameter2
    # ...
    # Optionally, return a value

# CALLING the function later in your code
function_name(argument1, argument2)

'''
```

**def**: The keyword to start a function definition.

**function_name**: A descriptive name you choose.

**parameters**: Variables that act as placeholders for the data the function needs. These are optional.

**arguments**: The actual values you pass to the function when you call it.

```python
# Example 1: A simple function with no parameters
def show_greeting():
    print("--------------------")
    print("Hello from a function!")
    print("--------------------")

# Now, let's call it multiple times
print("Before the first call.")
show_greeting()
print("In between calls.")
show_greeting()
print("After the second call.")


```

# ***The return Statement***

A function can perform an action (like print), but it can also send a value back to the code that called it. This is incredibly powerful.

**Real-life Analogy**: You ask a calculator (the function) to add 5 and 3 (the arguments). It doesn't just shout "8" to the room (print). It gives the value 8 back to you (the return value) so you can write it down or use it in another calculation.

```python
# Example 1: A function that calculates and returns a value
def add_two_numbers(a, b):
    result = a + b
    return result  # Sends the value of 'result' back

no1=int(input('Enter the first number : '))
no2=int(input('Enter the second number : '))
# Call the function and store the returned value in a variable
sum_of_numbers = add_two_numbers(no1, no2)
print(f"The sum is: {sum_of_numbers}")

# You can also use the return value directly
print(f"{no1} + {no2} is {add_two_numbers(no1, no2)}")


```

```python
# Example 2: Early exit with return
def check_password_strength(password):
    if len(password) < 8:
        # If the password is too short, return immediately.
        # The rest of the function will not run.
        return "Weak: Password must be at least 8 characters."

    if password.isnumeric():
        return "Weak: Password cannot be all numbers."

    # If it passes all checks...
    return "Strong"

# Test the function
pass1=input('Enter the first password : ')
pass2=input('Enter the second password : ')
feedback1 = check_password_strength(pass1)
feedback2 = check_password_strength(pass2)
print(f"Feedback for '{pass1}': {feedback1}")
print(f"Feedback for '{pass2}': {feedback2}")

```
